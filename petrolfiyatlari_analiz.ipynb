### Petrol Fiyatları Setiyle Örnek


```python
import pandas as pd
import matplotlib.pyplot as plt

# Gerekli modüllerin yükleme işlemleri
```

```python
gas = pd.read_csv("petrolfiyatlari.csv")
gas.head()

# CSV Uzantılı dosyayı okuma ve ilk 5 satırını ekrana yazdırma
# Okuttuğumuz dosya aynı dizinde olmalı aksi taktirde yolu belirtilmeli
```

```python
plt.title("Ülkelere Göre Petrol Fiyatları") # Grafiğin Başlığı
plt.plot(gas['Year'] , gas['USA'] , label = "USA" , color = "red" , linewidth = 2 , linestyle = "--" , marker = ".") # USA ' NIN GRAFİĞİNİ OLUŞTURUYORUM.'
plt.xlabel('Yıl')
plt.ylabel('Dolar')

#TABLOYA FRANSAYI'DA KOYUP ARALARINDAKİ İLİŞKİYİ İNCELİYECEĞİM

plt.plot(gas['Year'] , gas['France'] , label = "France" , color = "black" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.legend()
plt.show()



```

```python
# Bütün Ülkeleri Aynı tabloda toplamak istersek : 

plt.figure(figsize=(10,7)) #Tablonun boyutunu ayarladım. 

plt.title("Ülkelere Göre Petrol Fiyatları") # Grafiğin Başlığı
plt.plot(gas['Year'] , gas['USA'] , label = "USA" , color = "red" , linewidth = 2 , linestyle = "--" , marker = ".") # USA ' NIN GRAFİĞİNİ OLUŞTURUYORUM.'
plt.xlabel('Yıl')
plt.ylabel('Dolar')

#TABLOYA FRANSAYI'DA KOYUP ARALARINDAKİ İLİŞKİYİ İNCELİYECEĞİM

plt.plot(gas['Year'] , gas['France'] , label = "France" , color = "black" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.plot(gas['Year'] , gas['Australia'] , label = "Australia" , color = "yellow" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.plot(gas['Year'] , gas['Canada'] , label = "Canada" , color = "green" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.plot(gas['Year'] , gas['Germany'] , label = "Germany" , color = "brown" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.plot(gas['Year'] , gas['Italy'] , label = "Italy" , color = "pink" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.plot(gas['Year'] , gas['Japan'] , label = "Japan" , color = "blue" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.plot(gas['Year'] , gas['Mexico'] , label = "Mexico" , color = "gray" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.plot(gas['Year'] , gas['South Korea'] , label = "South Korea" , color = "orange" , linewidth = 2 , linestyle = "--" , marker = ".")
plt.plot(gas['Year'] , gas['UK'] , label = "UK" , color = "purple" , linewidth = 2 , linestyle = "--" , marker = ".")

plt.legend()
#Grafiği bilgisayarımıza kaydetmek için : 
plt.savefig("Yillara_gore_petrol_fiyatlari.png" , dpi = 300)
plt.show()
```
